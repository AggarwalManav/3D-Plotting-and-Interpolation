# 3-D Plotting 

## Line 

```python
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()
t = np.arange(0, 10*np.pi, np.pi/50)
x = np.sin(t)
y = np.cos(t)

ax.plot3D(x, y, t)
ax.set_title('3D Parametric Plot')

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('t', labelpad=20)

plt.show()
```

## Contour

```python
x=np.linspace(-5,5,11)
y=np.linspace(-5,5,11)
```

```python
x
```

```python
X,Y=np.meshgrid(x,y)
```

```python
#plotting parabolic function z
Z=X**2+Y**2
```

```python
fig=plt.figure()
plt.contourf(X,Y,Z,cmap="plasma")
plt.axis('scaled')
plt.colorbar()
plt.title("Contour 2D")
plt.savefig('contour2D.png')
plt.show()
```

## Surface and 3D Contour 

```python
fig=plt.figure()
ax=plt.axes(projection="3d")
ax.plot_surface(X,Y,Z,cmap="plasma",
                linewidth=0,
                antialiased=False,
                alpha=0.5)

ax.set_title('3D Surface Plot')

# Set axes label
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

#rotation
ax.view_init(30,60)
plt.savefig('3DplotSurface.png',dpi=600)
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.contour3D(X,Y,Z,100,cmap='binary')
ax.view_init(30,60)

ax.set_title('3D Contour Plot')

# Set axes label
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.savefig('3DplotContour.png',dpi=600)
plt.show()
```

# Linear Interpolation

```python
from scipy.interpolate import interp1d
```

```python
x=np.random.uniform(-20,20,size=50)
```

```python
def func(x):
    return x**2    
```

```python
y=func(x)
```

```python
newfunc=interp1d(x,y)
```

```python
x_new=np.linspace(-10,10,100)
```

```python
y_new=newfunc(x_new)
```

```python
plt.plot(x_new,func(x_new),label="original")
plt.plot(x_new,y_new,label="interpolated")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.figure(figsize=(3,3))
plt.savefig("linearInterpolation.png",dpi=600)
```

# Gridded Interpolation 

```python
from scipy.interpolate import interp2d
```

```python
def func(x,y):
    return (x+y)*np.exp(-5.0*(x**2+y**2))
```

```python
x=np.random.uniform(-1,1,size=50)
y=np.random.uniform(-1,1,size=50)
X,Y=np.meshgrid(x,y)
fvals=func(X,Y)
```

```python
fvals
```

```python
newfunc=interp2d(X,Y,fvals,kind="cubic")
```

```python
x_new=np.linspace(-1,1,100)
```

```python
y_new=x_new
```

```python
X_new,Y_new=np.meshgrid(x_new,y_new)
```

```python
f_new=newfunc(x_new,y_new)
```

```python
fvals.shape
```

```python
f_new.shape
```

```python
fig=plt.figure()
plt.contourf(X_new,Y_new,f_new,cmap="plasma")
plt.colorbar()
plt.axis("scaled")
plt.title("Interpolated Function Contour")
plt.savefig("InterpolatedFunctionContour.png")
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes(projection="3d")
ax.plot_surface(X_new,Y_new,f_new,cmap="plasma",
                linewidth=0,
                antialiased=False,
                alpha=0.5)
ax.set_title("Interpolated Function")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

#rotation
ax.view_init(20,100)
plt.savefig("InterpolatedFunction3D.png")
plt.show()
```

# Gradient Function

```python
dx=1
x=np.arange(0,6.1,dx)
print(x)
fx=x**2
print(fx)
dfdx=np.gradient(fx,dx)
print(dfdx)
```

```python
dx=0.25
x=np.arange(0,6.1,dx)
fx=x**2
dfdx=np.gradient(fx,dx)

print(x[:5])
print(fx[:5])
print(dfdx[:5])
dfdx_analytical=2*x
print(dfdx_analytical[:5])
```

### Gradient on 2D data

```python
data=np.array([
    [2,3,1,3,4],
    [4,3,4,6,5],
    [6,7,8,10,9],
    [3,4,6,7,8]
])
grad=np.gradient(data)
# default step size 1 for both axes
print(grad)
print(type(grad))
```

The first element of the list is partial derivative of data with respect to x-axis or 0-axis  
The second element of the list is partial derivative of data with respect to 
y-axis or 1-axis

```python
np.gradient(data,0.25,0.5)
#step value 0.25 in x-direction and 0.5 in y-direction
```

```python
np.gradient(data,0.25,axis=0)
```
